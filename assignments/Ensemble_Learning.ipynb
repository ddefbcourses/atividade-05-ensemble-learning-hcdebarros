# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `MNIST handwritten digits` (arquivos de treino e teste)
Realize a separação do conjunto de treino como treino e validação
Utilize `train_test_split` com controle de aleatoriedade (seed)
Retorne: `X_train`, `X_val`, `y_train`, `y_val`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

* Depende do modelo a ser utilizado. Se for Regressão Logística, SVM, KNN precisa normalizar, pois são sensíveis às escalas dos dados. Já Árvores de Decisão, Random Forest, nao precisa. Pois só fazem comparações entre os dados.

**Solução**:

```python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed):
    # Carregamento do dataset
    mnist = fetch_openml("mnist_784", version=1, as_frame=False)
    
    X = mnist.data    
    y = mnist.target  
    
    # 80% treino, 20% validação
    X_train, X_val, y_train, y_val = train_test_split(
        X, y, test_size=0.2, random_state=seed
    )
    
    return X_train, X_val, y_train, y_val
```

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

```python
from sklearn.ensemble import RandomForestClassifier

def train_random_forest(X_train, y_train, seed):
    # Criação do modelo
    modelo = RandomForestClassifier(random_state=seed)
    
    # Treino do modelo
    modelo.fit(X_train, y_train)
    
    return modelo
```

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

```python
from sklearn.metrics import accuracy_score

def evaluate(model, X_test, y_test):
    # Faz as predições
    y_pred = model.predict(X_test)
    
    # Calcula a acurácia
    acuracia = accuracy_score(y_test, y_pred)
    
    return acuracia
```

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

```python
def run_pipeline(model_type="rf", seed=42):
    # Carrega os dados
    X_train, X_val, y_train, y_val = load_data(seed)
    
    # Treina o modelo
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    
    # Avalia o modelo
    acuracia = evaluate(model, X_val, y_val)
    
    return acuracia

acuracia = run_pipeline(model_type="rf", seed=42)
print(acuracia)  
```

Execute o pipeline para ambos os modelos:

## Apresente:
- Acurácia, Precisão, Recall e F1-Score para o modelo

**Solução**:

```python
from sklearn.metrics import classification_report

def run_pipeline(model_type="rf", seed=42):
    # Carrega os dados
    X_train, X_val, y_train, y_val = load_data(seed)
    
    # Treina o modelo
    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    
    # Faz as predições
    y_pred = model.predict(X_val)
    
    # Mostra as métricas
    print(classification_report(y_val, y_pred))
    
    return model

run_pipeline(model_type="rf", seed=42)
```

# Questão 7

Para pelo menos um dos modelos:

- Compare a acurácia em treino e teste

## Responda:
- Existe overfitting?
* sim
- Qual modelo tende a sofrer mais com isso?
* Decision Tree, pois ela cria regras muito especificas para cada exemplo

```python
def avaliar_overfitting(model, X_train, y_train, X_val, y_val):
    acuracia_treino = evaluate(model, X_train, y_train)
    acuracia_val    = evaluate(model, X_val, y_val)

    print(f"Acurácia no Treino    : {acuracia_treino:.4f}")
    print(f"Acurácia na Validação : {acuracia_val:.4f}")
    print(f"Diferença             : {acuracia_treino - acuracia_val:.4f}")


X_train, X_val, y_train, y_val = load_data(seed=42)
model = train_random_forest(X_train, y_train, seed=42)
avaliar_overfitting(model, X_train, y_train, X_val, y_val)
```

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`

## Analise:
- O desempenho muda significativamente?
* No começo sim, depois não. O salto de 10 para 50 árvores melhora bastante. Mas de 100 para 200 a melhora é mínima.

```python
valores = [10, 50, 100, 200]

X_train, X_val, y_train, y_val = load_data(seed=42)

for n in valores:
    modelo = RandomForestClassifier(n_estimators=n, random_state=42)
    modelo.fit(X_train, y_train)
    acc = evaluate(modelo, X_val, y_val)
    print(f"n_estimators={n:>3} | Acurácia: {acc:.4f}")
```
